```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

path = './drive/MyDrive/ML2024/'
filename = 'drybean_scaled.csv'

data_pd = pd.read_csv(path + filename)
data = data_pd.values
X = data[:, :-1]
y = data[:, -1]

plt.scatter(X[:, 4], X[:, 12], c=y, cmap='viridis', s=50, alpha=0.7)
```

**K=2일 때 K-means clustering**

```python
from sklearn.cluster import KMeans

k=2
kmeans = KMeans(n_clusters=k, init='random', random_state=0)
kmeans.fit(X)

#클러스터링 이후 각 데이터의 군집 번호를 반환
labels = kmeans.labels_

#k개의 중심점을 반환
centroids = kmeans.cluster_centers_

# 각 클러스터의 SSE와 Total SSE 계산
total_sse = 0

for j in range(k):
    # j번 군집에 속하는 데이터 추출
    cluster_data = X[labels == j]

    # 군집마다의 SSE 계산
    sse = np.sum((cluster_data - centroids[j])**2)

    # Total SSE 계산
    total_sse += sse

print(f"K={k}일 때 SSE: {total_sse}")

plt.scatter(X[:, 4], X[:, 12], c=labels, cmap='viridis', s=50, alpha=0.7)
plt.scatter(centroids[:, 4], centroids[:, 12], c='red', s=100)
plt.show()
```
